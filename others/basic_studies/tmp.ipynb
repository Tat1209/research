```python
# このipynbのパスをもとに
from pathlib import Path
nb_path = Path("tmp.ipynb").resolve()
datasets_path = Path("/home/tat/research/assets/datasets")
# datasets_path = nb_path.parent
print(datasets_path)
```

```python
import torch
from torchvision.models import resnet18
from torchvision.datasets import CIFAR10

# このディレクトリにあるデータセットを取得．なかったらダウンロード
# root には，データセット本体ではなく，データセットが属するディレクトリを指定

train_ds = CIFAR10(root=datasets_path, train=True, download=True)
val_ds = CIFAR10(root=datasets_path, train=False, download=True)
model = resnet18(num_classes=100)
criterion = torch.nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.1)
```

```python
def train_1epoch(model, train_ds, criterion, optimizer):
    for inputs, labels in train_ds:
        # inputs, labels = inputs.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
```

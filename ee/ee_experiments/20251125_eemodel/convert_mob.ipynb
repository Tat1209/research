```python
import sys
from pathlib import Path

import torch
from torchvision.transforms import v2 as transforms

this_path = Path(__file__) if '__file__' in globals() else Path("<undefined>.ipynb").resolve()
work_path = next((p for p in this_path.parents if p.name == "research"), None)
tools_path = work_path / Path("../torch-tools")
sys.path.append(str(tools_path))

ee_tools_path_p = work_path / Path("ee")
sys.path.append(str(ee_tools_path_p))

from network import Network, Refiner
from trainer import Trainer
from datasets import fetch_handler

from torch import nn
```

```python
from torchvision.models import mobilenet_v2 as tv
from ee_tools.models.mobilenet_v2_ee import mobilenet_v2 as ee
from models.mobilenet_v2_cifar import mobilenet_v2 as cifar
from ee_tools.models.mobilenet_v2_cifar_ee import mobilenet_v2 as cifar_ee
# from torchvision.models import efficientnet_b0 as eff0
# from torchvision.models import convnext_base as cn

from ee_tools.ee_refiner import EERefiner
```

```python
input_size = (1, 3, 224, 224)
```

```python
base_network = tv(num_classes=100)
network = Network(base_network)
net_name = "tv_old"
p = Path(this_path.parent / f"{net_name}.txt")
ti = network.torchinfo(input_size=input_size)
p.write_text(str(ti))

base_network = tv(num_classes=100)
network = Network(EERefiner(base_network).build())
net_name = "tv_new"
p = Path(this_path.parent / f"{net_name}.txt")
ti = network.torchinfo(input_size=input_size)
p.write_text(str(ti))
```

```python
base_network = ee(num_classes=100, ee_div=2)
network = Network(base_network)
net_name = "ee_old"
p = Path(this_path.parent / f"{net_name}.txt")
ti = network.torchinfo(input_size=input_size)
p.write_text(str(ti))

base_network = tv(num_classes=100)
network = Network(EERefiner(base_network).multi_narrow(div=2).build())
net_name = "ee_new"
p = Path(this_path.parent / f"{net_name}.txt")
ti = network.torchinfo(input_size=input_size)
p.write_text(str(ti))

```

```python
base_network = cifar(num_classes=100)
network = Network(base_network)
net_name = "cifar_old"
p = Path(this_path.parent / f"{net_name}.txt")
ti = network.torchinfo(input_size=input_size)
p.write_text(str(ti))

base_network = tv(num_classes=100)
network = Network(EERefiner(base_network).cifar_style().build())
net_name = "cifar_new"
p = Path(this_path.parent / f"{net_name}.txt")
ti = network.torchinfo(input_size=input_size)
p.write_text(str(ti))
```

```python
base_network = cifar_ee(num_classes=100, ee_div=1)
network = Network(base_network)
net_name = "cifar_ee_old"
p = Path(this_path.parent / f"{net_name}.txt")
ti = network.torchinfo(input_size=input_size)
p.write_text(str(ti))

base_network = tv(num_classes=100)
network = Network(EERefiner(base_network).cifar_style().multi_narrow(div=8).build())
net_name = "cifar_ee_new"
p = Path(this_path.parent / f"{net_name}.txt")
ti = network.torchinfo(input_size=input_size)
p.write_text(str(ti))
```
